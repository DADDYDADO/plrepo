```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]        # 水平資料點
h = [10,20,30,40,50]   # 高度
plt.bar(x,h)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,]
h = [29,100,125,28]
color = ['b','orange','m','g']   # 顏色數據
label = ['calories(hundred)','protein','carbohydrate','lipid']   # 標籤數據
plt.bar(x,h,color=color,tick_label=label,width=0.5)  # 加入顏色、標籤和寬度參數
plt.show()
```
