```python
#beautifulsoup是python的函式庫模組
import requests
from bs4 import BeautifulSoup
```

```python
#將頁面的HTML抓取下來
#印出HTML
r = requests.get("https://www.ptt.cc/bbs/MuscleBeach/index.html")
print(r.text)
```

```python
#將網頁資料以html.parser存取下來
#取HTML標中的<div class="title"></div>中的<a>標籤存入sel
#目的是為了選取網頁裡的文章標題
soup = BeautifulSoup(r.text,"html.parser")
sel = soup.select("div.title a")
```

```python
#寫一個迴圈將所爬的印出來
for s in sel:
    print(s["href"], s.text)
```

```python
#擷取上一頁的網頁
u = soup.select("div.btn-group.btn-group-paging a")
url = "https://www.ptt.cc"+ u[1]["href"]
url
```

```python
#擷取多頁
url = "https://www.ptt.cc/bbs/MuscleBeach/index.html"
#往上爬4頁
for i in range(4): 
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    sel = soup.select("div.title a") #標題
    u = soup.select("div.btn-group.btn-group-paging a") #a標籤
    print ("本頁的URL為"+url)
    url = "https://www.ptt.cc"+ u[1]["href"] #上一頁的網址

    for s in sel: #印出網址跟標題
        print(s["href"],s.text)
```
