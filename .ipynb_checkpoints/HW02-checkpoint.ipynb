```python
import pandas as pd
songs = pd.read_csv("Streams.csv")
```

```python
# the last five songs of the list
songs.tail()
```

```python
import pandas as pd
features = pd.read_csv("Features.csv")
```

```python
features
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121)

# 繪製第一個圖表
STREAMS = songs["Streams (Billions)"]
ax1.hist(STREAMS)
ax1.set_title('Streams (Billions)')

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,6))
ax2 = fig.add_subplot(122)

# 繪製第二個圖表
KEYS = features["key"]
ax2.hist(KEYS)
ax2.set_title('Keys')

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 繪製第一個圖表
STREAMS = songs["Streams (Billions)"]
plt.hist(STREAMS, alpha=0.5, label='Streams (Billions)')

# 繪製第二個圖表
KEYS = features["key"]
plt.hist(KEYS, alpha=0.5, label='Keys')

# 設置圖表標題和標籤
plt.title('Streams and Keys Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# 顯示圖表
plt.show()
```

```python
#以上將兩個圖表整合，以顯示出'Streams'以及'Keys'兩者互相的關係圖
```
