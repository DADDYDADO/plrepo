```python
#beautifulsoup是python的函式庫模組
import requests
from bs4 import BeautifulSoup
```

```python
#將頁面的HTML抓取下來
#印出HTML
r = requests.get("https://www.ptt.cc/bbs/NBA/index.html")
print(r.text)
```

```python
#將網頁資料以html.parser存取下來
#取HTML標中的<div class="title"></div>中的<a>標籤存入sel
#目的是為了選取網頁裡的文章標題
soup = BeautifulSoup(r.text,"html.parser")
sel = soup.select("div.title a")
```

```python
#寫一個迴圈將所爬的印出來
for s in sel:
    print(s["href"], s.text)
```

```python
#擷取上一頁的網頁
u = soup.select("div.btn-group.btn-group-paging a")
url = "https://www.ptt.cc"+ u[1]["href"]
url
```

```python
#擷取多頁
url = "https://www.ptt.cc/bbs/NBA/index.html"
#往上爬4頁
for i in range(4): 
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    sel = soup.select("div.title a") #標題
    u = soup.select("div.btn-group.btn-group-paging a") #a標籤
    print ("本頁的URL為"+url)
    url = "https://www.ptt.cc"+ u[1]["href"] #上一頁的網址

    for s in sel: #印出網址跟標題
        print(s["href"],s.text)
```

```python
#利用python爬出圖片
import requests
```

```python
#變數名稱為pic
pic = requests.get('https://image-cdn.hypb.st/https%3A%2F%2Fhk.hypebeast.com%2Ffiles%2F2023%2F01%2Fhttps___hk.hypebeast.com_files_2021_08_never-gonna-give-you-up-passes-one-billion-views-01.jpg?fit=max&cbr=1&q=90&w=750&h=500')
```

```python
#將其轉為二進制方式
img2 = pic.content
img2
```

```python
pic_out = open('img1.png','wb') #img1.png為預存檔的圖片名稱
pic_out.write(img2) #將get圖片存入img1.png
pic_out.close() #關閉檔案(很重要)否則可能使文件保持打開的時間超過預期.
```
