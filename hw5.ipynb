```python
import pandas as pd
data = pd.read_csv("Seasons_Stats.csv")
```

```python
# the last five data of the list
data.tail()
```

```python
from tabulate import tabulate

# 筛选出"Year"列为2017的数据
filtered_data = data[data["Year"] == 2017]

# 根据"PTS"列对数据进行降序排序
sorted_data = filtered_data.sort_values(by="PTS", ascending=False)

# 获取前二十位得分最高的球员
top_20_players = sorted_data.head(20)[["Player","PTS"]]

# 打印结果
print(tabulate(top_20_players, headers='keys', tablefmt='psql'))
```

```python
import matplotlib.pyplot as plt

# 假设你已经将数据存储在名为df的Pandas数据帧中

# 筛选出"Year"列为2017的数据
filtered_data = data[data["Year"] == 2017]

# 根据"PTS"列对筛选后的数据进行降序排序
sorted_data = filtered_data.sort_values(by="PTS", ascending=False)

# 获取前二十位得分最高的球员的"Player"和"PTS"值
top_20_players = sorted_data.head(20)

# 提取球员名称和得分数据
players = top_20_players["Player"]
pts = top_20_players["PTS"]

# 设置图形大小
plt.figure(figsize=(8, 8))

# 绘制饼图
plt.pie(pts, labels=players, autopct='%1.1f%%')
plt.axis('equal')  # 使饼图为正圆形
plt.title("Top 20 Players' PTS in 2017")  # 添加标题
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 创建数据框
data = {
    'Player': ['Russell Westbrook', 'James Harden', 'Isaiah Thomas', 'Anthony Davis', 'Karl-Anthony Towns',
               'Damian Lillard', 'DeMar DeRozan', 'Stephen Curry', 'LeBron James', 'DeMarcus Cousins',
               'Andrew Wiggins', 'Kawhi Leonard', 'C.J. McCollum', 'Giannis Antetokounmpo', 'Kemba Walker',
               'Kyrie Irving', 'Jimmy Butler', 'John Wall', 'Bradley Beal', 'Paul George'],
    'PTS': [2558, 2356, 2199, 2099, 2061, 2024, 2020, 1999, 1954, 1942, 1933, 1888, 1837, 1832, 1830, 1816,
            1816, 1805, 1779, 1775]
}

df = pd.DataFrame(data)

# 重新索引以从最高得分到最低得分排序
df = df.sort_values(by='PTS', ascending=False).reset_index(drop=True)

# 创建热力图
fig, ax = plt.subplots()
im = ax.imshow([df['PTS']], cmap='hot')

# 添加轴标签
ax.set_xticks(range(len(df)))
ax.set_xticklabels(df['Player'], rotation=90)
ax.set_yticks([])  # 隐藏y轴刻度标签

# 添加数值标签
for i in range(len(df)):
    for j in range(len(df.columns)):
        text = ax.text(j, i, df.iloc[i, j],
                       ha="center", va="center", color="w")

# 添加颜色条
cbar = ax.figure.colorbar(im, ax=ax)

# 设置图表标题
ax.set_title("NBA Players - Points Scored")

# 显示图表
plt.show()
```
